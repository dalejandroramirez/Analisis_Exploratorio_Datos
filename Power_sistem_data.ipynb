```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
url="https://raw.githubusercontent.com/terranigmark/curso-analisis-exploratorio-datos-platzi/main/df_power.csv"
```

```python
data_power=pd.read_csv(url)
data_power
```

```python
data_power["Date"]=pd.to_datetime(data_power["Date"])
```

```python
data_power=data_power.set_index("Date")
```

```python
data_power["year"]=data_power.index.year
data_power["month"]=data_power.index.month
data_power["day"]=data_power.index.day

data_power
```

```python
import matplotlib.pyplot as plt
```

```python
data_power["Consumption"].plot(linewidth=0.3)
```

```python
name=["Consumption","Wind","Solar"]
data_power[name].plot(marker=".",alpha=0.5,figsize=(14,6),subplots=True)
```

Hacer un filtro que solo tome el 2016 y la columna llamada Consumption y luego graficarla

```python
data_power.loc["2016","Consumption"].plot()
```

Hacer un filtro que solo tome el 2016 en el mes de diciembre y la columna llamada Consumption y luego graficarla

```python
data_power.loc["2016-1":"2017-1","Consumption"].plot(marker="o")
```

```python
fig , axess = plt.subplots(3,1,figsize=(8,7))
for name ,ax in zip(["Consumption","Wind","Solar"],axess):
    sns.boxplot(data=data_power,x="month",y=name,ax=ax)
```
